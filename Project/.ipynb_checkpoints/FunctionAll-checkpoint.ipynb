```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from numpy import sqrt 
import time

#Tien Xu Ly
from sklearn.preprocessing import MinMaxScaler

#Draw Flot
import matplotlib.pyplot as plt

#Cacuale error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

#dtw Libaray
from dtw import dtw

#FFNN Libarary
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasRegressor
```

```python
#Xử lý dữ liệu thành dữ liệu đầu vào và đầu ra cho mô hình
def prepare_data(data, size_window,size_predict,stepWindow):
    X, y = [], []
    startWindow = 0
    for i in range(len(data) - size_window - 1):
        if (len(data[(startWindow + size_window):(startWindow + size_window + size_predict) , 0]) != size_predict):
            break
        X.append(data[startWindow:(startWindow + size_window), :])
        y.append(data[(startWindow + size_window):(startWindow + size_window + size_predict) , 0])
        print(data[startWindow:(startWindow + size_window), :])
        print("-----")
        print(data[(startWindow + size_window):(startWindow + size_window + size_predict) , 0])     
       
        startWindow += stepWindow
    return np.array(X), np.array(y)
```

```python
#Xử lý dữ liệu thành dữ liệu đầu vào và đầu ra cho mô hình
def prepare_data_mse(data, size_window,stepWindow):
    X= []
    startWindow = 0
    for i in range(len(data)):
        if (len(data[startWindow:(startWindow + size_window), :]) != size_window):
            break
        X.append(data[startWindow:(startWindow + size_window), :])
        startWindow += 1
    return np.array(X)
```

```python
# Fucntion Tính khoảng cách giữa 2 chuỗi thời gian
# ts1 chuỗi thứ nhất
# ts2 chuỗi thứ 2
def euclidean_distance(ts1, ts2):
    ts1= np.array(ts1)
    ts2= np.array(ts2)
    return np.linalg.norm(ts1-ts2)
```

```python
# function lấy ra k chuỗi gần nhất
# k số lượng chuỗi gần nhất
# distanceArr mảng khoảng cách
def kSimilarityTimeSeries(k, distanceArr):
    distances = np.array(distanceArr)
    return distances.argsort()[:k] 
```

```python
# Tính khoảng cách DTW
def dtw_dist(x, y):
    dist, _, _, _ = dtw(x, y, dist=lambda x, y: np.abs(x - y))
    return dist
```

```python
def addTestToTrain(X_train_KNN, y_train_KNN, xTest, yTest):
    X_train_KNN.append(xTest)
    y_train_KNN.append(yTest)
    return np.array(X_train_KNN), np.array(y_train_KNN)
```

```python
#Dự đoán
def predictDtw(k, X_train_KNN, y_train_KNN, X_test, y_test):
    y_pred_arr=[]
    for iTest in range(len(X_test)):
        if(k>len(X_train_KNN)):
            k=len(X_train_KNN)
        distanceArr=[]
        for iTrain in range(len(X_train_KNN)-size_window+2):
            distance = dtw_dist(X_test[iTest],X_train_KNN[iTrain])
            distanceArr.append(distance)
        indexKNN= kSimilarityTimeSeries(k,distanceArr)
        y_pred = np.mean(y_train_KNN[indexKNN])
        y_pred_arr.append(y_pred)
        X_train_KNN, y_train_KNN = addTestToTrain(X_train_KNN.tolist(), y_train_KNN.tolist(),X_test[iTest].tolist(), y_test[iTest].tolist())
        
    return np.array(y_pred_arr)
```

```python
def predictEculidean(k, X_train_KNN, y_train_KNN, X_test, y_test):
    y_pred_arr=[]
    for iTest in range(len(X_test)):
        if(k>len(X_train_KNN)):
            k=len(X_train_KNN)
        distanceArr=[]
        for iTrain in range(len(X_train_KNN)-size_window+2):
            distance = euclidean_distance(X_test[iTest],X_train_KNN[iTrain])
            distanceArr.append(distance)
        indexKNN= kSimilarityTimeSeries(k,distanceArr)
        y_pred = np.mean(y_train_KNN[indexKNN])
        y_pred_arr.append(y_pred)
        X_train_KNN, y_train_KNN = addTestToTrain(X_train_KNN.tolist(), y_train_KNN.tolist(),X_test[iTest].tolist(), y_test[iTest].tolist())
        
    return np.array(y_pred_arr)
```

```python
def caculateWeight(y_pred_FFNN,y_pred_KNN,y_test):
    FFNNSubKNN=[]
    TestSubKNN=[]
    weightEl=[]
    for i in range(len(y_pred_FFNN)):
        FFNNSubKNN.append(y_pred_FFNN[i]-y_pred_KNN[i])
        TestSubKNN.append(y_test[i]-y_pred_KNN[i])

    for j in range(len(FFNNSubKNN)):
        weightEl.append(((FFNNSubKNN[j]*TestSubKNN[j])/ (FFNNSubKNN[j]*FFNNSubKNN[j])))
    
    return np.array(weightEl)
```

```python

def combineKnnWithFFNN(n,best_k,isKnnPredictEculidean,best_neural,best_batch_size,best_epoch,filePath,size_window,size_predict,stepWindow):
    # Load dữ liệu
    dataCSV = pd.read_csv(filePath)
    dataFormat=dataCSV[['Close']]
    # Replace null values with 0
    dataFormat.fillna(0, inplace=True)
    # Remove outliers by replacing values outside of 15 standard deviations with the mean
    std = dataFormat['Close'].std()
    mean = dataFormat['Close'].mean()
    dataFormat['Close'] = np.where(dataFormat['Close'] > (mean + 10*std), mean, dataFormat['Close'])
    dataFormat['Close'] = np.where(dataFormat['Close'] < (mean - 10*std), mean, dataFormat['Close'])
    # # Scale data_AMZN to range [0, 1]
    scaler = MinMaxScaler()
    dataFormat['Close'] = scaler.fit_transform(dataFormat['Close'].values.reshape(-1, 1))
    
    # Fill in missing values with the mean of the previous and next values
    dataFormat['Close'] = dataFormat['Close'].interpolate(method='linear')
    data = dataFormat
    # Chia dữ liệu thành train set và test set
    train_size = int(len(data) * 0.8)
    train_data = data.iloc[:train_size, :]
    test_data = data.iloc[train_size:, :]
    X_train, y_train = prepare_data(train_data.values, size_window, size_predict, stepWindow)
    X_test, y_test = prepare_data(test_data.values, size_window, size_predict, stepWindow)
    if(isKnnPredictEculidean == 1):
        y_pred_KNN=predictEculidean(best_k, X_train, y_train, X_test, y_test)
    else:
        y_pred_KNN=predictDtw(best_k, X_train, y_train, X_test, y_test)
    # Tạo mô hình dự đoán lỗi MSE của KNN
    model_FFNN1 = Sequential()
    model_FFNN1.add(Dense(best_neural, input_dim=size_window, activation='sigmoid'))
    model_FFNN1.add(Dense(1))
    model_FFNN1.load_weights('bestWeight/ARNC_best_weights/best_weights_FFNN_'+str(best_neural)+'_Neural_'+str(best_batch_size)+'_BatchSize_'+str(best_epoch)+'_Epoch_ARNC.h5')
    model_FFNN1.compile(loss='mean_squared_error', optimizer='adam')
    y_pred_FFNN= model_FFNN1.predict(X_test)
    weight= caculateWeight(y_pred_FFNN,y_pred_KNN,y_test)
    y_pred_combine=[]
    for i in range(len(weight)):
        y_pred_combine.append(weight[i]*y_pred_FFNN[i]+(1-weight[i])*y_pred_KNN[i])
    y_pred_combine=np.array(y_pred_combine)
    mse_combine= mean_squared_error(y_test , y_pred_combine)
    return y_pred_combine,y_test
n=10
best_k=1
isKnnPredictEculidean= 1
best_neural=18
best_batch_size=8
best_epoch = 400
filePath= 'D:/DATN/Dataset/ARNC.csv'
size_window = 30
size_predict=1
stepWindow=15
y_pred_combine,y_test = combineKnnWithFFNN(n,best_k,isKnnPredictEculidean,best_neural,best_batch_size,best_epoch,filePath,size_window,size_predict,stepWindow)
```

```python
# Vẽ biểu đồ
plt.plot(y_test, label='Thực tế')
plt.plot(y_pred_combine, label='Dự đoán Combine')

plt.legend()
plt.show()
```

```python
def combineKnnWithFFNN(n,best_k,isKnnPredictEculidean,best_neural,best_batch_size,best_epoch,filePath,size_window,size_predict,stepWindow):
    # Load dữ liệu
    dataCSV = pd.read_csv(filePath)
    dataFormat=dataCSV[['Close']]
    # Replace null values with 0
    dataFormat.fillna(0, inplace=True)
    # Remove outliers by replacing values outside of 15 standard deviations with the mean
    std = dataFormat['Close'].std()
    mean = dataFormat['Close'].mean()
    dataFormat['Close'] = np.where(dataFormat['Close'] > (mean + n*std), mean, dataFormat['Close'])
    dataFormat['Close'] = np.where(dataFormat['Close'] < (mean - n*std), mean, dataFormat['Close'])
    # # Scale data_AMZN to range [0, 1]
    scaler = MinMaxScaler()
    dataFormat['Close'] = scaler.fit_transform(dataFormat['Close'].values.reshape(-1, 1))
    
    # Fill in missing values with the mean of the previous and next values
    dataFormat['Close'] = dataFormat['Close'].interpolate(method='linear')
    data = dataFormat
    # Chia dữ liệu thành train set và test set
    train_size = int(len(data) * 0.8)
    train_data = data.iloc[:train_size, :]
    test_data = data.iloc[train_size:, :]
    X_train, y_train = prepare_data(train_data.values, size_window, size_predict, stepWindow)
    X_test, y_test = prepare_data(test_data.values, size_window, size_predict, stepWindow)
    if(isKnnPredictEculidean == 1):
        y_pred_KNN=predictEculidean(best_k, X_train, y_train, X_test, y_test)
    else:
        y_pred_KNN=predictDtw(best_k, X_train, y_train, X_test, y_test)
    mseWithDay =[]
    for i in range(len(y_pred_KNN)):
        mseWithDay.append([np.mean((y_test[i] - y_pred_KNN[i])**2)])
    mseWithDay=np.array(mseWithDay)
    X_test_mse = prepare_data_mse(mseWithDay, size_window, stepWindow)
    # Tạo mô hình dự đoán lỗi MSE của KNN
    model_FFNN1 = Sequential()
    model_FFNN1.add(Dense(best_neural, input_dim=size_window, activation='sigmoid'))
    model_FFNN1.add(Dense(1))
    model_FFNN1.load_weights('bestWeight/ARNC_best_weights/best_weights_FFNN_'+str(best_neural)+'_Neural_'+str(best_batch_size)+'_BatchSize_'+str(best_epoch)+'_Epoch_ARNC.h5')
    model_FFNN1.compile(loss='mean_squared_error', optimizer='adam')
    y_pred_mse= model_FFNN1.predict(X_test_mse)
    y_pred_KNN_Split= y_pred_KNN[:len(y_pred_mse)]
    pred_knn_ffnn=[];
    for i in range(len(y_pred_mse)):
        pred_knn_ffnn.append(y_pred_KNN_Split[i]+y_pred_mse[i])
    pred_knn_ffnn=np.array(pred_knn_ffnn)
    return pred_knn_ffnn,(y_test[:len(pred_knn_ffnn)])
n=10
best_k=1
isKnnPredictEculidean= 1
best_neural=18
best_batch_size=8
best_epoch = 400
filePath= 'D:/DATN/Dataset/ARNC.csv'
size_window = 30
size_predict=1
stepWindow=15
pred_knn_ffnn,y_test = combineKnnWithFFNN(n,best_k,isKnnPredictEculidean,best_neural,best_batch_size,best_epoch,filePath,size_window,size_predict,stepWindow)
```
