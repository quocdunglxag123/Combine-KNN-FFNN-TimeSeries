$h_t = f(W_{hh} h_{t-1} + W_{xh} x_t + b_h)$

$h_t = f(W_{hh} h_{t-1} + W_{xh} x_t + b_h)$
Trong đó:

$h_t$ là trạng thái của nơ-ron đơn vị bộ nhớ tại thời điểm $t$

$h_{t-1}$ là trạng thái của nơ-ron đơn vị bộ nhớ tại thời điểm trước đó $t-1$

$x_t$ là đầu vào tại thời điểm $t$

$W_{hh}$ là ma trận trọng số liên kết giữa các nơ-ron đơn vị bộ nhớ

$W_{xh}$ là ma trận trọng số liên kết giữa đầu vào $x_t$ và nơ-ron đơn vị bộ nhớ

$b_h$ là vector độ lệch của nơ-ron đơn vị bộ nhớ

$f$ là hàm kích hoạt của nơ-ron đơn vị bộ nhớ

```python
data_KNN.shape
```

```python
# Replace null values with 0
data_KNN.fillna(0, inplace=True)

# Remove outliers by replacing values outside of 3 standard deviations with the mean
std_KNN = data_KNN['BTC'].std()
mean_KNN = data_KNN['BTC'].mean()
data_KNN['BTC'] = np.where(data_KNN['BTC'] > (mean_KNN + 3*std_KNN), mean_KNN, data_KNN['BTC'])
data_KNN['BTC'] = np.where(data_KNN['BTC'] < (mean_KNN- 3*std_KNN), mean_KNN, data_KNN['BTC'])

# # Scale data to range [0, 1]
scaler_KNN = MinMaxScaler()
data_KNN['BTC'] = scaler_KNN.fit_transform(data_KNN['BTC'].values.reshape(-1, 1))

# Fill in missing values with the mean of the previous and next values
data_KNN['BTC'] = data_KNN['BTC'].interpolate(method='linear')

# Print the cleaned data
print(data_KNN.head())
```

Thay thế các giá trị null bằng 0 .

Loại bỏ các giá trị nhiễu bằng cách thay thế các giá trị nằm ngoài khoảng 3 độ lệch chuẩn của dữ liệu với giá trị trung bình của dữ liệu.

MinMaxScaler để chuyển đổi dữ liệu thành phạm vi [0, 1]. Tiếp theo, chúng ta sử dụng 

Interpolate để điền vào các giá trị bị thiếu bằng cách lấy trung bình của các giá trị trước và sau đó.


std() là một hàm trong Python để tính độ lệch chuẩn của một tập dữ liệu. 

Công thức tính độ lệch chuẩn của một tập dữ liệu là:
σ = sqrt(1/n * Σ(xi - x_mean)^2)

Ví dụ, nếu chúng ta có tập dữ liệu [2, 4, 6, 8], thì giá trị trung bình là 5, độ lệch của mỗi giá trị so với giá trị trung bình lần lượt là [-3, -1, 1, 3], giá trị bình phương độ lệch của mỗi giá trị so với giá trị trung bình lần lượt là [9, 1, 1, 9], đưa vào công thức độ lệch chuẩn
 ==> σ = sqrt(1/4 * (9 + 1 + 1 + 9)) = sqrt(20/4) = sqrt(5) ≈ 2.236

```python
#Xử lý dữ liệu thành dữ liệu đầu vào và đầu ra cho mô hình
def prepare_data_KNN(data_KNN, size_window_KNN,size_predict_KNN,stepWindow_KNN):
    X_KNN, y_KNN = [], []
    startWindow_KNN = 0
    for i in range(len(data_KNN) - size_window_KNN - 1):
        if (len(data_KNN[(startWindow_KNN + size_window_KNN):(startWindow_KNN + size_window_KNN + size_predict_KNN) , 0]) != size_predict_KNN):
            break
        X_KNN.append(data_KNN[startWindow_KNN:(startWindow_KNN + size_window_KNN), :])
        y_KNN.append(data_KNN[(startWindow_KNN + size_window_KNN):(startWindow_KNN + size_window_KNN + size_predict_KNN) , 0])
        print(data_KNN[startWindow_KNN:(startWindow_KNN + size_window_KNN), :])
        print("-----")
        print(data_KNN[(startWindow_KNN + size_window_KNN):(startWindow_KNN + size_window_KNN + size_predict_KNN) , 0])     
       
        startWindow_KNN += stepWindow_KNN
    return np.array(X_KNN), np.array(y_KNN)

size_window_KNN = 20
size_predict_KNN=1
stepWindow_KNN=1
X_train_KNN, y_train_KNN = prepare_data_KNN(train_data_KNN.values, size_window_KNN, size_predict_KNN, stepWindow_KNN)
X_test_KNN, y_test_KNN = prepare_data_KNN(test_data_KNN.values, size_window_KNN, size_predict_KNN, stepWindow_KNN)
```

```python
# Tính khoảng cách DTW
def dtw_dist(x, y):
    dist, _, _, _ = dtw(x, y, dist=lambda x, y: np.abs(x - y))
    return dist

# Xây dựng mô hình KNN với DTW là khoảng cách
knn_dtw = KNeighborsRegressor(n_neighbors=5, metric=dtw_dist)
```

```python
# Huấn luyện mô hình
knn_dtw.fit(X_train_KNN.reshape(X_train_KNN.shape[0], -1), y_train_KNN)
```

```python
# Dự đoán kết quả
y_pred_KNN = knn_dtw.predict(X_test_KNN.reshape(X_test_KNN.shape[0], -1))
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasRegressor
```

```python
data_FFNN = pd.read_csv('D:/DATN/Dataset/portfolio_data.csv')
data_FFNN=data_FFNN[['BTC']]
data_FFNN
```

```python
# Replace null values with 0
data_FFNN.fillna(0, inplace=True)

# Remove outliers by replacing values outside of 3 standard deviations with the mean
std_FFNN = data_FFNN['BTC'].std()
mean_FFNN = data_FFNN['BTC'].mean()
data_FFNN['BTC'] = np.where(data_FFNN['BTC'] > (mean_FFNN + 3*std_FFNN), mean_FFNN, data_FFNN['BTC'])
data_FFNN['BTC'] = np.where(data_FFNN['BTC'] < (mean_FFNN - 3*std_FFNN), mean_FFNN, data_FFNN['BTC'])

# # Scale data to range [0, 1]
scaler_FFNN = MinMaxScaler()
data_FFNN['BTC'] = scaler_FFNN.fit_transform(data_FFNN['BTC'].values.reshape(-1, 1))

# Fill in missing values with the mean of the previous and next values
data_FFNN['BTC'] = data_FFNN['BTC'].interpolate(method='linear')

# Print the cleaned data
print(data_FFNN.head())
```

```python
# Chia dữ liệu thành train set và test set
train_size_FFNN = int(len(data_FFNN) * 0.8)
train_data_FFNN = data_FFNN.iloc[:train_size_FFNN, :]
test_data_FFNN = data_FFNN.iloc[train_size_FFNN:, :]
```

```python
#Xử lý dữ liệu thành dữ liệu đầu vào và đầu ra cho mô hình
def prepare_data_FFNN(data_FFNN, size_window_FFNN, size_predict_FFNN, stepWindow_FFNN):
    X_FFNN, y_FFNN = [], []
    startWindow_FFNN = 0
    for i in range(len(data_FFNN) - size_window_FFNN - 1):
        if (len(data_FFNN[(startWindow_FFNN + size_window_FFNN):(startWindow_FFNN + size_window_FFNN + size_predict_FFNN) , 0]) != size_predict_FFNN):
            break
        X_FFNN.append(data_FFNN[startWindow_FFNN:(startWindow_FFNN + size_window_FFNN), :])
        y_FFNN.append(data_FFNN[(startWindow_FFNN + size_window_FFNN):(startWindow_FFNN + size_window_FFNN + size_predict_FFNN) , 0])
        print(data_FFNN[startWindow_FFNN:(startWindow_FFNN + size_window_FFNN), :])
        print("-----")
        print(data_FFNN[(startWindow_FFNN + size_window_FFNN):(startWindow_FFNN + size_window_FFNN + size_predict_FFNN) , 0])     
       
        startWindow_FFNN += stepWindow_FFNN
    return np.array(X_FFNN), np.array(y_FFNN)

size_window_FFNN = 20
size_predict_FFNN=1
stepWindow_FFNN=1
X_train_FFNN, y_train_FFNN = prepare_data_FFNN(train_data_FFNN.values, size_window_FFNN, size_predict_FFNN, stepWindow_FFNN)
X_test_FFNN, y_test_FFNN = prepare_data_FFNN(test_data_FFNN.values, size_window_FFNN, size_predict_FFNN, stepWindow_FFNN)
```

```python
# create the model
def create_model_FFNN():
    model_FFNN = Sequential()
    model_FFNN.add(Dense(10, input_dim=size_window_FFNN, activation='sigmoid'))
    model_FFNN.add(Dense(1))
    model_FFNN.compile(loss='mean_squared_error', optimizer='adam')
    return model_FFNN

# create the model
model_FFNN = KerasRegressor(build_fn=create_model_FFNN, verbose=0)
```

```python
#define the grid search parameters
batch_size_FFNN = [16, 32, 64]
epochs_FFNN = [50, 100, 200]
param_grid_FFNN = dict(batch_size=batch_size_FFNN, epochs=epochs_FFNN)
```

```python
# perform the grid search
grid_FFNN = GridSearchCV(estimator=model_FFNN, param_grid=param_grid_FFNN, cv=3)
grid_result_FFNN = grid_FFNN.fit(X_train_FFNN, y_train_FFNN)
```

```python
# print the best results
print("Best: %f using %s" % (grid_result_FFNN.best_score_, grid_result_FFNN.best_params_))
```

```python
# train the model with the best parameters
best_params_FFNN = grid_result_FFNN.best_params_
model_FFNN = create_model_FFNN()
model_FFNN.fit(X_train_FFNN, y_train_FFNN, epochs=best_params_FFNN['epochs'], batch_size=best_params_FFNN['batch_size'], verbose=2, callbacks=[EarlyStopping(monitor='loss', patience=10)], shuffle=False)
```

```python
# make predictions on the test set
predictions_FFNN = model_FFNN.predict(X_test_FFNN)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from dtw import dtw

# Load dữ liệu
data = pd.read_csv('long_data_.csv')

# Chia dữ liệu thành train set và test set
train_size = int(len(data) * 0.8)
train_data = data.iloc[:train_size, :]
test_data = data.iloc[train_size:, :]

# Xử lý dữ liệu thành dữ liệu đầu vào và đầu ra cho mô hình
def prepare_data(data, time_steps):
    X, y = [], []
    for i in range(len(data) - time_steps - 1):
        X.append(data[i:(i + time_steps), :])
        y.append(data[i + time_steps, 0])
    return np.array(X), np.array(y)

time_steps = 20
X_train, y_train = prepare_data(train_data.values, time_steps)
X_test, y_test = prepare_data(test_data.values, time_steps)

# Tính khoảng cách DTW
def dtw_dist(x, y):
    dist, _, _, _ = dtw(x, y, dist=lambda x, y: np.abs(x - y))
    return dist

# Xây dựng mô hình KNN với DTW là khoảng cách
knn_dtw = KNeighborsRegressor(n_neighbors=5, metric=dtw_dist)

# Huấn luyện mô hình
knn_dtw.fit(X_train.reshape(X_train.shape[0], -1), y_train)

# Dự đoán kết quả
y_pred = knn_dtw.predict(X_test.reshape(X_test.shape[0], -1))

# Tính MSE
mse = np.mean(np.square(y_test - y_pred))
print('MSE:', mse)

# Vẽ biểu đồ
plt.plot(y_test, label='Thực tế')
plt.plot(y_pred, label='Dự đoán')
plt.legend()
plt.show()
```
