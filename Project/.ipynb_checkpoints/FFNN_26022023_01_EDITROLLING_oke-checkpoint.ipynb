```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
```

```python
data  = pd.read_csv('D:/DATN/Dataset/archive/long_data_.csv')
```

```python
# Replace null values with 0
data.fillna(0, inplace=True)

# Remove outliers by replacing values outside of 3 standard deviations with the mean
std = data['Usage'].std()
mean = data['Usage'].mean()
data['Usage'] = np.where(data['Usage'] > (mean + 3*std), mean, data['Usage'])
data['Usage'] = np.where(data['Usage'] < (mean - 3*std), mean, data['Usage'])

# Scale data to range [0, 1]
scaler = MinMaxScaler()
data['Usage'] = scaler.fit_transform(data['Usage'].values.reshape(-1, 1))

# Fill in missing values with the mean of the previous and next values
data['Usage'] = data['Usage'].interpolate(method='linear')

# Print the cleaned data
print(data.head())
```

```python
data = data.groupby(data['Dates'], as_index=False).mean()
data =data [['Usage']]
data 
```

```python
# Chia dữ liệu thành train set và test set
train_size = int(len(data) * 0.8)
train_data = data.iloc[:train_size, :]
test_data = data.iloc[train_size:, :]
```

```python
#Xử lý dữ liệu thành dữ liệu đầu vào và đầu ra cho mô hình
def prepare_data(data, size_window,size_predict,stepWindow):
    X, y = [], []
    startWindow = 0
    for i in range(len(data) - size_window - 1):
        if (len(data[(startWindow + size_window):(startWindow + size_window + size_predict) , 0]) != size_predict):
            break
        X.append(data[startWindow:(startWindow + size_window), :])
        y.append(data[(startWindow + size_window):(startWindow + size_window + size_predict) , 0])
        print(data[startWindow:(startWindow + size_window), :])
        print("-----")
        print(data[(startWindow + size_window):(startWindow + size_window + size_predict) , 0])     
       
        startWindow += stepWindow
    return np.array(X), np.array(y)

size_window = 20
size_predict=1
stepWindow=1
X_train, y_train = prepare_data(train_data.values, size_window,size_predict,stepWindow)
X_test, y_test = prepare_data(test_data.values, size_window,size_predict,stepWindow)
```

```python
# Khởi tạo mô hình
model = Sequential()
model.add(Dense(50, activation='sigmoid', input_dim=X_train.shape[1]))
model.add(Dense(1))
```

```python
# Huấn luyện mô hình
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
# Huấn luyện mô hình
model.fit(X_train, y_train, epochs=100, batch_size=16, verbose=1)
```

```python
# Dự đoán kết quả 
y_pred_scaled = model.predict(X_test)
```

```python
mse = np.mean(np.square(y_test - y_pred_scaled))
print('MSE:', mse)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred_scaled)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred_scaled)
r2
```

```python
# chuyển về đơn vị thực tế
y_pred = scaler.inverse_transform(y_pred_scaled)
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python
mport pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense

# Load dữ liệu
data = pd.read_csv('long_data_.csv')

# Chia dữ liệu thành train set và test set
train_size = int(len(data) * 0.8)
train_data = data.iloc[:train_size, :]
test_data = data.iloc[train_size:, :]

# Chuẩn hóa dữ liệu
scaler = MinMaxScaler()
train_data_scaled = scaler.fit_transform(train_data)
test_data_scaled = scaler.transform(test_data)

# Xử lý dữ liệu thành dữ liệu đầu vào và đầu ra cho mô hình
def prepare_data(data, time_steps):
    X, y = [], []
    for i in range(len(data) - time_steps - 1):
        X.append(data[i:(i + time_steps), :])
        y.append(data[i + time_steps, 0])
    return np.array(X), np.array(y)

time_steps = 20
X_train, y_train = prepare_data(train_data_scaled, time_steps)
X_test, y_test = prepare_data(test_data_scaled, time_steps)

# Khởi tạo mô hình
model = Sequential()
model.add(Dense(50, activation='sigmoid', input_dim=X_train.shape[1]))
model.add(Dense(1))

# Compile mô hình
model.compile(loss='mean_squared_error', optimizer='adam')

# Huấn luyện mô hình
model.fit(X_train, y_train, epochs=100, batch_size=16, verbose=1)

# Dự đoán kết quả và chuyển về đơn vị thực tế
y_pred_scaled = model.predict(X_test)
y_pred = scaler.inverse_transform(y_pred_scaled)
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))

# Tính MSE
mse = np.mean(np.square(y_test - y_pred))
print('MSE:', mse)

# Vẽ biểu đồ
plt.plot(y_test, label='Thực tế')
plt.plot(y_pred, label='Dự đoán')
plt.legend()
plt.show()
```
