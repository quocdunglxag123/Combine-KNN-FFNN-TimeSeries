```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from numpy import sqrt 

#Tien Xu Ly
from sklearn.preprocessing import MinMaxScaler

#Draw Flot
import matplotlib.pyplot as plt

#Cacuale error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

#KNN Libaray
from sklearn.neighbors import KNeighborsRegressor
from dtw import dtw

#FFNN Libarary
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasRegressor
```

```python
# Load dữ liệu
dataCSV = pd.read_csv('D:/DATN/Dataset/portfolio_data.csv')
dataCSV
```

```python
data_AMZN=dataCSV[['AMZN']]
data_DPZ=dataCSV[['DPZ']]
data_BTC=dataCSV[['BTC']]
data_NFLX=dataCSV[['NFLX']]
```

```python
# Replace null values with 0
data_AMZN.fillna(0, inplace=True)

# Remove outliers by replacing values outside of 3 standard deviations with the mean
std = data_AMZN['AMZN'].std()
mean = data_AMZN['AMZN'].mean()
data_AMZN['AMZN'] = np.where(data_AMZN['AMZN'] > (mean + 3*std), mean, data_AMZN['AMZN'])
data_AMZN['AMZN'] = np.where(data_AMZN['AMZN'] < (mean - 3*std), mean, data_AMZN['AMZN'])

# # Scale data_AMZN to range [0, 1]
scaler = MinMaxScaler()
data_AMZN['AMZN'] = scaler.fit_transform(data_AMZN['AMZN'].values.reshape(-1, 1))

# Fill in missing values with the mean of the previous and next values
data_AMZN['AMZN'] = data_AMZN['AMZN'].interpolate(method='linear')

# Print the cleaned data_AMZN
print(data_AMZN)
```

```python
# Replace null values with 0
data_DPZ.fillna(0, inplace=True)

# Remove outliers by replacing values outside of 3 standard deviations with the mean
std = data_DPZ['DPZ'].std()
mean = data_DPZ['DPZ'].mean()
data_DPZ['DPZ'] = np.where(data_DPZ['DPZ'] > (mean + 3*std), mean, data_DPZ['DPZ'])
data_DPZ['DPZ'] = np.where(data_DPZ['DPZ'] < (mean - 3*std), mean, data_DPZ['DPZ'])

# # Scale data_DPZ to range [0, 1]
scaler = MinMaxScaler()
data_DPZ['DPZ'] = scaler.fit_transform(data_DPZ['DPZ'].values.reshape(-1, 1))

# Fill in missing values with the mean of the previous and next values
data_DPZ['DPZ'] = data_DPZ['DPZ'].interpolate(method='linear')

# Print the cleaned data_DPZ
print(data_DPZ)
```

```python
# Replace null values with 0
data_BTC.fillna(0, inplace=True)

# Remove outliers by replacing values outside of 3 standard deviations with the mean
std = data_BTC['BTC'].std()
mean = data_BTC['BTC'].mean()
data_BTC['BTC'] = np.where(data_BTC['BTC'] > (mean + 3*std), mean, data_BTC['BTC'])
data_BTC['BTC'] = np.where(data_BTC['BTC'] < (mean - 3*std), mean, data_BTC['BTC'])

# # Scale data_BTC to range [0, 1]
scaler = MinMaxScaler()
data_BTC['BTC'] = scaler.fit_transform(data_BTC['BTC'].values.reshape(-1, 1))

# Fill in missing values with the mean of the previous and next values
data_BTC['BTC'] = data_BTC['BTC'].interpolate(method='linear')

# Print the cleaned data_BTC
print(data_BTC)
```

```python
# Replace null values with 0
data_NFLX.fillna(0, inplace=True)

# Remove outliers by replacing values outside of 3 standard deviations with the mean
std = data_NFLX['NFLX'].std()
mean = data_NFLX['NFLX'].mean()
data_NFLX['NFLX'] = np.where(data_NFLX['NFLX'] > (mean + 3*std), mean, data_NFLX['NFLX'])
data_NFLX['NFLX'] = np.where(data_NFLX['NFLX'] < (mean - 3*std), mean, data_NFLX['NFLX'])

# # Scale data_NFLX to range [0, 1]
scaler = MinMaxScaler()
data_NFLX['NFLX'] = scaler.fit_transform(data_NFLX['NFLX'].values.reshape(-1, 1))

# Fill in missing values with the mean of the previous and next values
data_NFLX['NFLX'] = data_NFLX['NFLX'].interpolate(method='linear')

# Print the cleaned data_NFLX
print(data_NFLX)
```

Thay thế các giá trị null bằng 0 .

Loại bỏ các giá trị nhiễu bằng cách thay thế các giá trị nằm ngoài khoảng 3 độ lệch chuẩn của dữ liệu với giá trị trung bình của dữ liệu.

MinMaxScaler để chuyển đổi dữ liệu thành phạm vi [0, 1]. Tiếp theo, chúng ta sử dụng 

Interpolate để điền vào các giá trị bị thiếu bằng cách lấy trung bình của các giá trị trước và sau đó.


std() là một hàm trong Python để tính độ lệch chuẩn của một tập dữ liệu. 

Công thức tính độ lệch chuẩn của một tập dữ liệu là:
σ = sqrt(1/n * Σ(xi - x_mean)^2)

Ví dụ, nếu chúng ta có tập dữ liệu [2, 4, 6, 8], thì giá trị trung bình là 5, độ lệch của mỗi giá trị so với giá trị trung bình lần lượt là [-3, -1, 1, 3], giá trị bình phương độ lệch của mỗi giá trị so với giá trị trung bình lần lượt là [9, 1, 1, 9], đưa vào công thức độ lệch chuẩn
 ==> σ = sqrt(1/4 * (9 + 1 + 1 + 9)) = sqrt(20/4) = sqrt(5) ≈ 2.236

```python
df_MseWith_Neural = pd.DataFrame([],  columns =  ["Neural", "MSE"])
for neural in range(5,21):
    # create the model
    def create_model_FFNN():
        model_FFNN = Sequential()
        model_FFNN.add(Dense(neural, input_dim=size_window, activation='sigmoid'))
        model_FFNN.add(Dense(1))
        model_FFNN.compile(loss='mean_squared_error', optimizer='adam')
        return model_FFNN
    # create the model
    model_FFNN = KerasRegressor(build_fn=create_model_FFNN, verbose=0)
    #define the grid search parameters
    batch_size_FFNN = [8, 16, 32, 64, 128]
    epochs_FFNN = [50, 100, 150, 200, 250, 300, 350, 400]
    param_grid_FFNN = dict(batch_size=batch_size_FFNN, epochs=epochs_FFNN)
    # perform the grid search
    grid_FFNN = GridSearchCV(estimator=model_FFNN, param_grid=param_grid_FFNN, cv=3)
    grid_result_FFNN = grid_FFNN.fit(X_train, y_train)
    # print the best results
    print("Best: %f using %s" % (grid_result_FFNN.best_score_, grid_result_FFNN.best_params_))
    # train the model with the best parameters
    best_params_FFNN = grid_result_FFNN.best_params_
    model_FFNN = create_model_FFNN()
    model_FFNN.fit(X_train, y_train, epochs=best_params_FFNN['epochs'], batch_size=best_params_FFNN['batch_size'], verbose=2, callbacks=[EarlyStopping(monitor='loss', patience=10)], shuffle=False)
    # make predictions on the test set
    predictions_FFNN = model_FFNN.predict(X_test)
    # get the best weights from the model
    best_weights_FFNN = model_FFNN.get_weights()
    print("Best weights:", best_weights_FFNN)
    # create the model
    model_FFNN1 = Sequential()
    model_FFNN1.add(Dense(neural, input_dim=size_window, activation='sigmoid'))
    model_FFNN1.add(Dense(1))
    model_FFNN1.set_weights(best_weights_FFNN)
    model_FFNN1.compile(loss='mean_squared_error', optimizer='adam')
    model_FFNN1.set_weights(best_weights_FFNN)
    model_FFNN1.fit(X_train, y_train, epochs=best_params_FFNN['epochs'], batch_size=best_params_FFNN['batch_size'], verbose=1)
    y_pred= model_FFNN1.predict(X_test)
    rowMseWithNeural = pd.Series([neural,mean_squared_error(y_test , y_pred)], index=df_MseWith_Neural.columns)
    df_MseWith_Neural = df_MseWith_Neural.append(rowMseWithNeural,ignore_index=True)
df_MseWith_Neural = df_MseWith_Neural.sort_values('MSE')
best_Neural=df_MseWith_Neural['Neural'].iloc[0]   
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from dtw import dtw

# Load dữ liệu
data = pd.read_csv('long_data_.csv')

# Chia dữ liệu thành train set và test set
train_size = int(len(data) * 0.8)
train_data = data.iloc[:train_size, :]
test_data = data.iloc[train_size:, :]

# Xử lý dữ liệu thành dữ liệu đầu vào và đầu ra cho mô hình
def prepare_data(data, time_steps):
    X, y = [], []
    for i in range(len(data) - time_steps - 1):
        X.append(data[i:(i + time_steps), :])
        y.append(data[i + time_steps, 0])
    return np.array(X), np.array(y)

time_steps = 20
X_train, y_train = prepare_data(train_data.values, time_steps)
X_test, y_test = prepare_data(test_data.values, time_steps)

# Tính khoảng cách DTW
def dtw_dist(x, y):
    dist, _, _, _ = dtw(x, y, dist=lambda x, y: np.abs(x - y))
    return dist

# Xây dựng mô hình KNN với DTW là khoảng cách
knn_dtw = KNeighborsRegressor(n_neighbors=5, metric=dtw_dist)

# Huấn luyện mô hình
knn_dtw.fit(X_train.reshape(X_train.shape[0], -1), y_train)

# Dự đoán kết quả
y_pred = knn_dtw.predict(X_test.reshape(X_test.shape[0], -1))

# Tính MSE
mse = np.mean(np.square(y_test - y_pred))
print('MSE:', mse)

# Vẽ biểu đồ
plt.plot(y_test, label='Thực tế')
plt.plot(y_pred, label='Dự đoán')
plt.legend()
plt.show()
```
