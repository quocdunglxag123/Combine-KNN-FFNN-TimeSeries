```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from numpy import sqrt 

#Tien Xu Ly
from sklearn.preprocessing import MinMaxScaler

#Draw Flot
import matplotlib.pyplot as plt

#Cacuale error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

#KNN Libaray
from sklearn.neighbors import KNeighborsRegressor
from dtw import dtw

#FFNN Libarary
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasRegressor
```

```python
#param KNN: 
best_k=2
#param FFNN
best_neural=9
size_window = 20
step_window=1
```

```python
# Load dữ liệu
dataCSV = pd.read_csv('D:/DATN/Dataset/portfolio_data.csv')
dataCSV
```

```python
data_BTC=dataCSV[['BTC']]
```

```python
data_BTC
```

```python
data = data_BTC
```

```python
scaler_KNN = MinMaxScaler()
data['BTC'] = scaler_KNN.fit_transform(data['BTC'].values.reshape(-1, 1))
```

```python
data
```

```python
# Chia dữ liệu thành train set và test set
train_size = int(len(data) * 0.8)
train_data = data.iloc[:train_size, :]
test_data = data.iloc[train_size:, :]
```

```python
# Xử lý dữ liệu thành dữ liệu đầu vào và đầu ra cho mô hình
def prepare_data(data, time_steps):
    X, y = [], []
    for i in range(len(data) - time_steps - 1):
        X.append(data[i:(i + time_steps), :])
        y.append(data[i + time_steps, 0])
    return np.array(X), np.array(y)
X_train, y_train = prepare_data(train_data.values, size_window)
X_test, y_test = prepare_data(test_data.values, size_window)
```

```python
def eucli(ts1, ts2):
    ts1= np.array(ts1)
    ts2= np.array(ts2)
    return np.linalg.norm(ts1-ts2)

# Xây dựng mô hình KNN với DTW là khoảng cách
knn_dtw = KNeighborsRegressor(n_neighbors=5, metric=eucli)
# Huấn luyện mô hình
knn_dtw.fit(X_train.reshape(X_train.shape[0], -1), y_train)
# Dự đoán kết quả
y_pred_KNN = knn_dtw.predict(X_test.reshape(X_test.shape[0], -1))
```

```python
y_pred_KNN
```

```python
mse_KNN= mean_squared_error(y_test , y_pred_KNN)
mse_KNN
```

```python
# Vẽ biểu đồ
plt.plot(y_test, label='Thực tế')
plt.plot(y_pred_KNN, label='Dự đoán')
plt.legend()
plt.show()
```

```python
mseWithDay =[]
for i in range(len(y_pred_KNN)):
    mseWithDay.append([np.mean((y_test[i] - y_pred_KNN[i])**2)])
mseWithDay=np.array(mseWithDay)
mseWithDay
```

```python
len(mseWithDay)
```

```python
#Xử lý dữ liệu thành dữ liệu đầu vào và đầu ra cho mô hình
def prepare_data_mse(data, size_window,stepWindow):
    X= []
    startWindow = 0
    for i in range(len(data)):
        if (len(data[startWindow:(startWindow + size_window), :]) != size_window):
            break
        X.append(data[startWindow:(startWindow + size_window), :])
        startWindow += stepWindow
    return np.array(X)
X_test_mse = prepare_data_mse(mseWithDay, size_window, step_window)
```

```python
# vì rolling nên kích thước mảng cửa sổ bằng chiều dài tập dữ - chiều dài của sổ + 1
# ví dụ có 11 điểm dữ liệu, chiều dài cửa sổ = 4 thì sẽ có 11-4+1= 8 cửa sổ
len(X_test_mse)
```

```python
X_test_mse
```

```python
# Huấn luyện mô hình với param tốt nhất để get lại best_weight
model_FFNN = Sequential()
model_FFNN.add(Dense(best_neural, input_dim=size_window))
model_FFNN.add(Dense(1))
model_FFNN.compile(loss='mean_squared_error', optimizer='adam')
model_FFNN.fit(X_train, y_train)
best_weights_FFNN = model_FFNN.get_weights()
```

```python
# Tạo mô hình dự đoán lỗi MSE của KNN
model_FFNN1 = Sequential()
model_FFNN1.add(Dense(best_neural, input_dim=size_window))
model_FFNN1.add(Dense(1))
model_FFNN1.set_weights(best_weights_FFNN)
model_FFNN1.compile(loss='mean_squared_error', optimizer='adam')
model_FFNN1.fit(X_train, y_train)
y_pred_mse= model_FFNN1.predict(X_test_mse)
```
